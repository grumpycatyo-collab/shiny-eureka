# Machine Learning

```python
#to remember, here is our cleaned out dataset:
base
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

le.fit(base["generation_name"])
base["Gen"]=le.transform(base["generation_name"])

le.fit(base["mark"])
base["Mark"]=le.transform(base["mark"])

le.fit(base["model"])
base["Model"]=le.transform(base["model"])
base = base.drop(columns=['model','mark','generation_name'])
base
```

We tokenize all string values to make the life of the model easier. Next we will try some models, and after that deduce what's the best one.

```python
X = base.drop(columns="Mark")           
y = base["Mark"]
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25)
print("X Train : ", X_train.shape)
print("X Test  : ", X_test.shape)
print("Y Train : ", y_train.shape)
print("Y Test  : ", y_test.shape)
```

Sizes of arrays, to be clear. Now let's assume that Linear Regression will be the less accurate and the Decision Tree model will be the most accurate one (because of it's algorithm).

## Linear Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
linear = LinearRegression().fit(X_train, y_train)
pred_LR = linear.predict(X_test)
print("Score with X_train & y_train is: ",linear.score(X_train,y_train))
print("Score of success with X_test & y_test is: ",linear.score(X_test,y_test))
print("Mean squared error: %.2f" % mean_squared_error(y_test, pred_LR))
```

## Ridge Regression

```python
from sklearn import linear_model
ridge = linear_model.Ridge(alpha=.9).fit(X_train, y_train)
pred_ridge = ridge.predict(X_test)

print("Score with X_train & y_train is: ",ridge.score(X_train,y_train))
print("Score of success with X_test & y_test is: ",ridge.score(X_test,y_test))
print("Mean squared error: %.2f" % mean_squared_error(y_test, pred_ridge))
```

## K Neighbours Regression

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=1,weights='uniform',algorithm='auto',leaf_size=30,\
    p=2, metric='minkowski', metric_params=None, n_jobs=None).fit(X_train, y_train)
pred_neigh = neigh.predict(X_test)
ui = neigh.score(X_test,y_test)
print("Score with X_train & y_train is: ",neigh.score(X_train,y_train))
print("Score of success with X_test & y_test is: ",ui)
print("Mean squared error: %.2f" % mean_squared_error(y_test, pred_neigh))
```

## Decision Tree Classifier

```python
dtc = tree.DecisionTreeClassifier().fit(X_train,y_train)
pred_dtc = dtc.predict(X_test)
print("Score with X_train & y_train is: ",dtc.score(X_train,y_train))
print("Score of success with X_test & y_test is: ",dtc.score(X_test,y_test))
print("Mean squared error: %.2f" % mean_squared_error(y_test, pred_dtc))
```

## Conclusions

```python
print("Scores for all models:")
print("Linear Regression Model: ",linear.score(X_test,y_test))
print("Decision Tree Classifier: ",dtc.score(X_test,y_test))
print("K Neighbors Regressor: ",ui)
print("Ridge Regression Model: ",ridge.score(X_test,y_test))
```
